### Resampling

The wavelength dispersion of `paintbox` objects can be resampled to a new dispersion with the `Resample` operator.  

```python
import numpy as np
import paintbox as pb
import matplotlib.pyplot as plt

wave = np.linspace(5000, 8000, 10000)
order = 5
poly = pb.Polynomial(wave, order)
# New wavelength range with two segments
new_wave = np.hstack([np.linspace(5500, 6200, 20), np.linspace(6700, 7800, 20)])
new_poly = pb.Resample(new_wave, poly)
theta = np.random.random(order + 1)
fig = plt.figure(figsize=(8, 6))
plt.plot(wave, poly(theta))
plt.plot(new_wave, new_poly(theta), "x")
plt.xlabel("$\lambda$ (Angstrom)")
plt.ylabel("Flux")
plt.show()
```

The resampling is performed with [spectres](https://spectres.readthedocs.io/en/latest/), thus it can handle arbitrary wavelength values within the limits of the original wavelength dispersion. Values outside the original range are set to zero.
